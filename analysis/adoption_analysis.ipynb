```python
# load the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# load the data
data = pd.read_csv('pet_export.csv')
data.head()
```

```python
# segregate user features from the pet swipe
user_features = data[['houseType', 'petsVaccinated','familySupport','allergies','petsNeutered', 'liveAlone', 'fullname', 'currentPetTypes', 'petsAllowed', 'about', 'timestamp', 'hasPets','childrenBelow12','uid']]
print(user_features.head())

# all other columns are pet swipe
pet_swipe = data.drop(['Unnamed: 0', 'houseType', 'petsVaccinated','familySupport','allergies','petsNeutered', 'liveAlone', 'currentPetTypes', 'petsAllowed', 'about', 'timestamp', 'hasPets','childrenBelow12','uid'], axis=1)
print(pet_swipe.head())
```

```python
# count number of cell with value of 0 in each row
pet_swipe['dislike_count'] = pet_swipe.apply(lambda x: (x == 0).sum(), axis=1)

# count number of cell with value of 1 in each row
pet_swipe['like_count'] = pet_swipe.apply(lambda x: (x == 1).sum(), axis=1)

# all users with 0 like_count and sort by dislike_count
users = pet_swipe[(pet_swipe['like_count'] == 0) & (pet_swipe['dislike_count'] > 0)].sort_values(by='dislike_count', ascending=False)
print(users)
```

```python
# count the number of 1 per column
pet_likes = {}
for col in pet_swipe.columns:
    if col!='fullname':
        pet_likes[col] = pet_swipe[col].sum()

# sort the dictionary by value
sorted_pet_likes = dict(sorted(pet_likes.items(), key=lambda item: item[1], reverse=True))

print(sorted_pet_likes)
```

```python
# calculate the score for each user
for user_index in user_features.index:
    user_score = 0
    user = user_features.loc[user_index]
    if user['houseType'] == 'house with backyard':
        user_score += 1
    if user['houseType'] == 'townhouse':
        user_score += 1
    if user['petsVaccinated'] == 'yes':
        user_score += 1
    if user['familySupport'] == 'no':
        user_score -= 1
    if user['petsAllowed'] == 'no':
        user_score -= 3
    if user['allergies'] == 'yes':
        user_score -= 3
    if user['petsNeutered'] == 'no':
        user_score -= 1
    if user['childrenBelow12'] == 'yes':
        user_score -= 1
    user_features.loc[user_index, 'score'] = user_score
# sort the users by score
sorted_users = user_features.sort_values(by='score', ascending=False)
display(sorted_users)
        
```
